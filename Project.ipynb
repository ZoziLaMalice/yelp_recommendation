```
import pyspark.sql.functions as F

display(
    user_cuisine.groupBy("UserID").count().orderBy(F.desc('count'))
)
```

```
display(
    cuisine.groupBy("PlaceID").count().orderBy(F.desc('count'))
)
```

```
all_cuisine = (cuisine.join(user_cuisine, 'Rcuisine', 'inner'))

display(all_cuisine)
```

```
prospects = (cuisine.join(user_cuisine, 'Rcuisine', 'inner')
                           .groupBy('PlaceID')
                           .agg(F.collect_set('UserID').alias('users'))
                           .withColumn('count_users', F.size('users'))
                           .orderBy(F.desc('count_users')))

display(prospects)
```

```
display(
hours.withColumn('day', F.explode( F.split( F.col('days').substr(F.lit(0), F.length(F.col('days')) - 1), ';') ))
  .groupBy('placeID')
  .pivot('day')
  .agg(F.first('hours'))
  .select('placeID', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun')
)
```

```
place_rating = place.join(rating,"placeID")
display(place_rating)
```

```
import pyspark.sql.functions as f

mean_ratings = place_rating.groupBy('placeID').agg(f.mean('rating')\
     .alias('mean_rating'),f.count('rating')\
     .alias('number_ratings'))

display(mean_ratings)
```

# NPS

```
display(rating.groupBy('placeID')
       .pivot('rating')
       .count()
       )
```

```
from pyspark.sql.window import Window

display(
  rating.groupBy('placeID')
  .pivot('rating')
  .count()
  .fillna(0)
  .withColumn('Total', F.col('0') + F.col('1') + F.col('2'))
  .withColumn('NPS', (F.col('2') / F.col('Total') - F.col('0') / F.col('Total')))
  .join(cuisine, 'placeID', 'inner')
  .withColumn('Global Rank', F.rank().over(Window.orderBy(F.desc('NPS'))))
  .withColumn('Rank', F.rank().over(Window.partitionBy('Rcuisine').orderBy(F.desc('NPS'))))
  .orderBy('Global Rank')
  .join(prospects, 'placeID', 'left')
             )
```

# Recommandations

## 1 - Preprocessing

```
%sh pip install mlflow
```

```
from pyspark.ml.feature import StringIndexer

userIdIndexer = StringIndexer(inputCol="userID", outputCol="userIdIndex").fit(rating)

rating = userIdIndexer.transform(rating)

display(rating)
```

```
from pyspark.sql.types import *
from pyspark.sql import functions as F

rating = rating.select(
  F.col("userIdIndex"),
  F.col("placeID").cast(IntegerType()),
  F.col("rating").cast(IntegerType())
)

display(rating)
```

## 2 - Entrainement

```
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import BinaryClassificationEvaluator, RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

train, test = rating.randomSplit([0.8, 0.2])
```

```
als = ALS(userCol="userIdIndex", itemCol="placeID", ratingCol="rating", coldStartStrategy="drop", nonnegative=True)


evaluator = RegressionEvaluator(metricName="rmse", labelCol="rating", predictionCol="prediction")

paramGrid = ParamGridBuilder() \
                    .addGrid(als.rank, [1, 5, 10]) \
                    .addGrid(als.maxIter,[1,5,10]) \
                    .build()

cv = CrossValidator(
  estimator=als, 
  evaluator=evaluator, 
  estimatorParamMaps=paramGrid, 
  numFolds=5
)
```

```
model = cv.fit(train)
```

```
bestModel = model.bestModel

bestModelParameters = {
  key.name: value
  for key, value in bestModel.extractParamMap().items()
  if key.name in ["rank", "maxIter"]
}
```

## 3 - Evaluation

```
preds_test = model.transform(test)
```

```
evaluator.evaluate(preds_test)
```

```
display(preds_test)
```
